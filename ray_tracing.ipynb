```python
import numpy as np
import pandas as pd
```

```python
def val(theta, phi):
    '''
     Theta is elevation and phi is azimuth
     x is horizontal, y is vertical and z is distance
     theta is -pi/2 to pi/2
     phi is from -pi to pi
    '''
    
    x = np.sin(theta)*np.cos(phi)
    y = np.sin(theta)*np.sin(phi)
    z = np.cos(theta)
    
    col = ['X', 'Y', 'Z']
    coordinate = pd.DataFrame(columns=col)
    
    for dr in range(1,50,1):
        coordinate.loc[dr, 'X'] = x*dr
        coordinate.loc[dr, 'Y'] = y*dr
        coordinate.loc[dr, 'Z'] = z*dr
    
    return coordinate
        
        
def planeEquation(a, b, c):
    '''
    Equation of plane ax + by + cz + d
    '''
    
    x1 = a[0]
    y1 = a[1]
    z1 = a[2]
    x2 = b[0]
    y2 = b[1]
    z3 = b[2]
    x3 = c[0]
    y3 = c[1]
    z3 = c[2]
    
    
    vector1 = [x2 - x1, y2 - y1, z2 - z1]
    vector2 = [x3 - x1, y3 - y1, z3 - z1]

    cross_product = [vector1[1] * vector2[2] - vector1[2] * vector2[1],\
                     -1 * (vector1[0] * vector2[2] - vector1[2] * vector2[0]), \
                     vector1[0] * vector2[1] - vector1[1] * vector2[0]]

    a = cross_product[0]
    b = cross_product[1]
    c = cross_product[2]
    d = - (cross_product[0] * x1 + cross_product[1] * y1 + cross_product[2] * z1)
    
    return a,b,c,d
```

```python
coordinates = val(np.pi/6, np.pi/3)
```

```python
pointA = np.array([1,2,-5])
pointB = np.array([2,5,-9])
pointC = np.array([2,4,-5])
a,b,c,d = planeEquation(pointA,pointB,pointC)
print a,b,c,d
```

```python
den = (a**2+b**2+c**2)**0.5
```

```python
for i in range(1,50):
    num = a * coordinates.loc[i,'X'] + b * coordinates.loc[i, 'Y'] + c * coordinates.loc[i, 'Z'] + d
    val = num/den
    print val
#     if val < 0.1 and val > -0.1:
#         print 'index :', i, ', val :', float(val)
#         break
```
